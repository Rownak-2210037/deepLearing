# 1.RNN(Recurrent Neural Network)
RNN stands for Recurrent Neural Network - a type of artificial neural network designed to work with sequential data.It has memory and remembers previous inputs while processing new ones.
Think of it like this:

Regular neural networks: Look at each piece of data independently.

RNNs: Remember previous information while processing new data.

### 1.1 Why Do We Need RNN?

Normal Neural Network:
Treats every input independently and no memory of previous data

But in real life:

Sentence meaning depends on previous words.

Stock price depends on previous days,

Speech depends on previous sound.

### 1.2 How RNN Works:
Simple RNN Structure:
    
    Output 
         |
    [Hidden State]  (Memory from previous step)
         
    New Input

At each time step:

ht=f(W*x(t)+Uh(t-1))

Where: 

X(t)= current input

h(t-1)= previous hidden state (memory)

ht= new hidden state

It keeps updating memory step-by-step.

### 1.3 Where RNN is Used?
-> Language modeling:Google Translate

-> Text Prediction:Next word suggestion

-> Speech recognition:Voice to text

-> Time series prediction

-> Chatbots

-> Stock Market:Price prediction

**Based on Internal Structure**

1 Simple RNN:

-> Basic RNN with short memory

-> Passes information through each step

-> Struggles with long sequences

-> Has vanishing gradient problem




2 LSTM (Long Short-Term Memory):

-> Advanced RNN with long memory

-> Has special gates to remember/forget selectively

-> Perfect for long sequences

LSTM Recurrent Neural Network is a special version of the RNN model. It stands for Long Short-Term Memory. The simple RNN has a problem that it cannot remember the context in a long sentence because it quickly loses information. And that is why Simple RNN has an only short-term memory.

LSTM has both long-term and short-term memory. It can store any contextual information for a long time. 

LSTM has 2 internal states. 

1. Memory Cell State which acts as a long term memory.

2. Hidden State which acts as a short term memory

3. Hidden state = the part of the memory the model uses right now to make predictions, while cell state keeps all important memory for later.

The main working components in LSTM are gates. There are 3 types of gates in LSTM:

1. Forget Gate:removes unnecessary old memory.

2. Input Gate:adds important new memory.

3. Output Gate:decides what memory to use for output.LSTM learns from data which parts of memory are important. The gates weights are trained to keep, forget, or use information so that predictions are accurate.





3 GRU (Gated Recurrent Unit):

-> Simplified LSTM (faster but still good memory)

-> Combines remember/forget gates into one

-> Good balance of speed and memory


So, we have to Choose wisely such that:

Short text/speech  Simple RNN

Complex long sequences  LSTM

Need speed + memory  GRU






Cell state stores long-term information, while hidden state stores short-term information and helps generate output.





### 1.10 Why Long Sentences Are a Problem in RNN?
Vanishing Gradient Problem.

RNN uses only hidden state, so it can forget long-term info and make mistakes on long sequences.

But we use RNN when sequences are short because its simpler, faster, and sufficient for small tasks.Such as short sentence POS tagging, short time series prediction, small character-level tasks.Sequence is short, hidden state doesnt lose important info.

Sentence:

"I grew up in France many years ago ... now I speak fluent French."

To predict "French", the model must remember "France" from far back.

In a simple RNN:

-> Each step multiplies weights again and again.

-> During backpropagation, gradients become very small.

-> Only hidden state

-> No protection for memory

-> Repeated multiplication makes gradients very small

=> So it forgets long-term information

-> Early information slowly fades away.So by the time we reach:

"I speak fluent ___"

The model may forget "France".

In RNN:

ht=f(W*x(t)+Uh(t-1))

The hidden state keeps multiplying through many time steps.

If values are:

   Less than 1  they shrink  vanishing gradient

   Greater than 1  they explode  exploding gradient

For long sentences:
  Memory becomes weak.

  Important early words disappear.

**Again**

In normal RNN,Memory becomes weak for long sequences.

ht=f(W*x(t)+Uh(t-1))

Each time step:

New hidden state depends on previous hidden state

This happens again and again for many words

So for a long sentence:

h  h  h  ...  h

Information from h must pass through 99 steps.

During this process:

It keeps getting multiplied by weights.

Because of repeated multiplication.

Imagine:

Weight value = 0.5

After many steps:

0.5  0.5  0.5  0.5  ... (many times)

It becomes:

0.00001 (very small)

So early information becomes almost zero.

That means the model forgets early words.



**What Is Vanishing Gradient Problem?**

During training, we use backpropagation through time.

Gradients flow backward through many steps.

If gradient value is:

< 1  it shrinks every step

After many steps  becomes almost 0

This is called Vanishing Gradient

When gradient  0:

Weights do not update properly

Model cannot learn long-term dependency


**Why LSTM Solves This?**

LSTM adds:

Cell state (separate memory highway)

Gates to control memory

Less repeated shrinking

So important information is protected.






### 1.11 LSTM inside what happened:












Output gate only decides what to select, but hidden state is the activated, usable version of that selection.

We cannot directly use o alone because its just a filter (0s and 1s), not the actual memory values.










### 1.11 Word Embedding in RNN
Neural networks cannot understand text like:

"I love AI"

They only understand numbers.So we must convert words into numbers.One-Hot Encoding is an approach for converting words into numbers but it has some limitations.Words embeddings mean vector representation os words in number.

I    = [1,0,0]

love = [0,1,0]

AI   = [0,0,1]

Problem:

-> No meaning between words

-> "love" and "like" look completely different

-> Very large vector if vocab is big




















```python
import torch
import torch.nn as nn

# Vocabulary
vocab = {"I":0, "love":1, "AI":2}
vocab_size = len(vocab)

class RNNModel(nn.Module):
    def __init__(self):
        super(RNNModel, self).__init__()
        
        self.embedding = nn.Embedding(
            num_embeddings=vocab_size,
            embedding_dim=5   # each word  5 numbers
        )
        
        self.rnn = nn.RNN(
            input_size=5,
            hidden_size=6,
            batch_first=True
        )
        
        self.fc = nn.Linear(6, vocab_size)

    def forward(self, x):
        x = self.embedding(x)   # word  vector
        out, hidden = self.rnn(x)
        last_out = out[:, -1, :]
        output = self.fc(last_out)
        return output


model = RNNModel()

# Input: "I love"
x = torch.tensor([[0,1]])

output = model(x)
print(output)
```

## Code:Basic RNN Code using pytorch.

```python
import torch
import torch.nn as nn

texts=[
  "i love python",
  "pytorch is amazing",
  "deep learning is fun" 
]

# step-1: tokenization(actual splitting)

tokenized_texts=[text.split() for text in texts]   # tokenized_texts=[["i","love","python"],["pytorch","is","amazing"],["deep","learning","is","fun"]] list of list of strings
print(tokenized_texts)
print("\n")

# step-2: build vocabulary(Vocabulary = Word  ID Mapper)

words=set()
for tokens in tokenized_texts:
    words.update(tokens)         # add tokens to the set  words = {'i', 'love', 'python', 'pytorch', 'is', 'amazing', 'deep', 'learning', 'fun'}  stores unique value

vocab= {}  #ALWAYS an empty DICTIONARY
# Assign index to each word (dictionary)
for i,word in enumerate(words):     # "enumerate() gives BOTH the index AND the item in each loop, so we don't need a separate counter!" 
  vocab[word]=i
print(vocab)                     # {'i': 0, 'love': 1, 'python': 2, 'pytorch': 3, 'is': 4, 'amazing': 5, 'deep': 6, 'learning': 7, 'fun': 8}-->  dictionary of word and index
print("\n")
inv_vocab = {i: word for word, i in vocab.items()}
vocab_size = len(vocab)


# step-3: numericalization (word  number)

numerical_texts = [] 
for tokens in tokenized_texts: 
  numerical_texts.append([vocab[token] for token in tokens])
print(numerical_texts)     # [[0, 1, 2], [3, 4, 5], [6, 7, 4, 8]] list of list of numbers
#  Words converted to numbers
```

```python
# step-4: make input and target 
# Creates MULTIPLE training pairs from each sentence!
# More data = better learning! creates MANY small lessons from each sentence, teaching the model to predict at EVERY position, not just the end

inputs=[]
targets=[]
for sentence in numerical_texts:
    for i in range(len(sentence)-1):
        inputs.append(sentence[:i+1])  # input = all words up to the current word
        targets.append(sentence[i+1])   # target = the next word
        
print("inputs:", inputs)   # inputs: [[0], [0, 1], [3], [3, 4], [6], [6, 7], [6, 7, 4]]
print("targets:", targets) # targets: [1, 2, 4, 5, 7, 4, 8]       
```

```python
# padding Convert to tensors Python lists are for humans, tensors are for PyTorch models! Tensors are like multi-dimensional arrays that PyTorch can efficiently process.
# If all sequences have the SAME length  NO padding needed! If lengths are DIFFERENT  NEED padding!
from torch.nn.utils.rnn import pad_sequence

# Convert and pad in one go
x = pad_sequence([torch.tensor(seq) for seq in inputs], batch_first=True)
y = torch.tensor(targets)

print(x.shape)  # torch.Size([7, 3])   7 sequences, each length 3
print(y.shape) 
```

```python
x
```

```python
# step-5: define RNN model architecture(Embedding  RNN  Linear)

class NextWordRNN(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_size):
        super().__init__()
        
        self.embedding = nn.Embedding(vocab_size, embed_dim)       # embedding_dim = 4 Each word described by 4 numbers

        self.rnn = nn.RNN(embed_dim, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, vocab_size)
        
    def forward(self, x):
        x = self.embedding(x)              # (batch, seq_len)  (batch, seq_len, embed_dim)      # word  vector
        out, h_n = self.rnn(x)             # (batch, seq_len, hidden)                            # out contains hidden states for EVERY time step:
        last_hidden = out[:, -1, :]  
        # take last time step and because for next-word prediction,we want the summary after reading full input.
        out = self.fc(last_hidden)         # (batch, vocab_size)    # # Convert hidden states to vocabulary-sized prediction(scores for each word in vocabulary)
        return out
```

```python
# step-6: create model

model = NextWordRNN(vocab_size, embed_dim=10, hidden_size=16)

criterion = nn.CrossEntropyLoss()        # loss function for multi-class classification
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)   # optimizer updates model parameters based on computed gradients and learning rate

# Adam(Adaptive Moment Estimation) is the SMART GPS that remembers where it's been and adjusts its step size to find the lowest point (best model) faster
```

```python
# step-7: training loop (Forward pass  Compute loss  Backward pass  Update parameters)

for epoch in range(200):
    
    optimizer.zero_grad()
    
    outputs = model(x)
    loss = criterion(outputs, y)
    
    loss.backward()
    optimizer.step()
    
    if epoch % 50 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```

```python
# STEP 6: Predict Next Word

test_sentence = "pytorch is"
test_encoded = torch.tensor([[vocab[word] for word in test_sentence.split()]])

with torch.no_grad():
    prediction = model(test_encoded)
    predicted_index = torch.argmax(prediction, dim=1).item()     # torch.argmax()--> Selects highest score word.

print("Next word:", inv_vocab[predicted_index])
```

### 2.Long-Short Term Memory process:
Firstly, Data preparation(unsup-supervised,words-numbers,embaddings)

Secondly,model architecture

Thirdly, training loop

Finally, prediction


1 Start with raw text and tokenize the sentences into words.

2 Build a vocabulary and convert each word into an index (word  number).

3 Convert sentences into sequences of indices and apply padding so all sequences have equal length.

4 Pass these sequences into an Embedding layer to convert indices into dense word vectors.

5 Feed the embeddings into an LSTM layer to process the sequence and capture context.

6 Take the final hidden state, pass it through a Linear layer (and Softmax/Sigmoid) to make predictions, then train using loss + backpropagation.

### 2.1 Why embedding?

 To give words meaningful dense vectors instead of meaningless IDs.Embedding = converting something into a meaningful vector representation.

Numbers = Just IDs (like roll numbers) - no meaning!

Embeddings = Smart numbers that capture meaning and relationships!

apple  = 1

banana = 2

orange = 3

king   = 4

queen  = 5

man    = 6

woman  = 7

Problem: 
- apple(1) and banana(2) are close in numbers (1,2)  good for fruits
  
- BUT king(4) and queen(5) also close (4,5)  but 4-5 = 1, 1-2 = 1
  
- Computer thinks: "apple is as related to banana as king is to queen?"
  
- Worse: apple(1) and king(4) distance = 3, same as orange(3) and queen(5) distance = 2

- MEANINGLESS! Numbers don't capture relationships!


Suppose vocabulary:

I  0

love  1

AI  2

If we send these directly into LSTM:

[0, 1, 2]

Model thinks:

2 > 1 > 0

AI > love > I

But words do NOT have numeric meaning.

Numbers are just IDs.


**One-hot solution**

I     [1,0,0], love  [0,1,0], AI    [0,0,1]

Problems:

-> Very large vector if vocab is big

-> No semantic meaning

-> Sparse (mostly zeros)

Each word = many numbers (a vector):

- apple  banana (both fruits)

### 2.2 Forward pass in LSTM:

Sentence:

"I love AI"

Step-by-step:

1 Convert words  indices

2 Embedding converts  vectors

3 LSTM processes step by step:

At each time step:

(h_t, C_t) = LSTM(x_t, h_{t-1}, C_{t-1})

Meaning:

Uses previous memory (h, C)

Uses current word vector

Updates memory

After last word:

Final hidden state  Linear  Prediction

That is forward pass.


### 2.3 Backward pass in LSTM:

After prediction:

1 Compare prediction with true target

2 Compute loss

3 Compute gradients

4 Backpropagation Through Time (BPTT)

Gradients flow:

Output-> Linear-> Last hidden-> Previous hidden->Embedding

Weights are updated:

Embedding weights

LSTM weights

Linear weights

So embedding also learns better word vectors.




### 2.4 Why Embedding is Better than One-Hot?

love  [0.8, 0.6, -0.2]

like  [0.75, 0.55, -0.1]

Vectors are similar  model understands similarity.

3 One-Hot is Huge

If vocabulary = 10,000 words

One-hot vector size = 10,000

Mostly zeros  waste of memory

Embedding size might be 100  small and efficient 

4 Embedding Learns Automatically

Embedding vectors improve during training.

```python
# step-5:Convert words to indices using the vocabulary
# Neural networks cannot understand words.They can only process numbers.This step converts words  indices that we can feed into embeddings / LSTM
def text_to_indices(sentence,vocab):
    numerical_sentence=[]
    for token in sentence:
        if token in vocab:
            numerical_sentence.append(vocab[token]) # convert word to index using vocab
        else:
            numerical_sentence.append(vocab['<unk>']) # if word not in vocab, use index of <unk>
    return numerical_sentence
```

```python
# step-9: Now we have x and y ready for training the model. We can create a custom dataset and dataloader to feed this data into the model during training.

class CustomDataset(Dataset):
    def __init__(self, x, y):
        self.x = x
        self.y = y               # Store data properly in the dataset class so that it can be accessed during training.

    def __len__(self):
        return len(self.x)       # Return the number of samples in the dataset. This is required for PyTorch to know how many batches to create during training.

    def __getitem__(self, idx):  # This method is called by the DataLoader to get a single sample of data. It takes an index (idx) as input and returns the corresponding input and target sequences from x and y.
        return self.x[idx], self.y[idx]
    
# Why do we need Dataset?

# We use a class because PyTorchs training system is built to work with Dataset objects that define how to get and count data samples.
# When training a model:

# We have input data (x)

# We have target labels (y)

# We dont feed everything at once

# We feed data in small batches

# PyTorch needs a Dataset object to handle this properly.
```

**LSTM & Embedding Basic Summary**


1 Why Embedding Uses 100 Dimensions?

Each word index is just an ID (no meaning).

Embedding converts each word into a vector of 100 numbers.

These numbers represent hidden semantic features.

100 is a hyperparameter (can be 50, 200, 300, etc.).

2 Are Embedding Values Random?

Yes, initially they are random.

During training, they are updated using backpropagation.

Words appearing in similar contexts become similar vectors.

Embedding layer is a trainable lookup table.

3 What Does nn.LSTM(100, 150) Mean?

100 = input size (embedding dimension).

150 = hidden size (memory vector size).

LSTM reads a 100-dim word vector and produces a 150-dim hidden state.

Hidden state stores contextual information.

4 What is Hidden Size (150)?

It is the size of LSTMs internal memory (h_t and c_t).

Each time step outputs a 150-dimensional vector.

Larger hidden size = more learning capacity.

It does NOT mean number of words.

Suppose we represent words like this:

"cat" = 1
"dog" = 2
"car" = 3

Problem 

Model thinks:

dog (2) is closer to car (3)

cat (1) is far from car (3)

But numbers 1,2,3 have no real meaning.

Word index is just an ID  not meaning.

5 What is Embedding Actually?

Instead of 1 number, we give each word a vector.

Example (small size 3):

cat  = [0.9, 0.1, 0.2]

dog  = [0.85, 0.15, 0.25]

car  = [0.1, 0.8, 0.9]

Now:

cat and dog vectors look similar.

car looks different.

Now model understands semantic similarity.

During training.

When we train the model:

loss.backward()

optimizer.step()

The embedding values get updated using gradients.

```python
# step-10: embedding layer with vector (32,61) to (32.61.100) 32 sentences in a batch, each sentence has 61 words (after padding), and we want to convert each word into a 100-dimensional vector using an embedding layer.

class LSTMModel(nn.Module):             # creating a neural network model.It inherits from PyTorchs base model class.
    def __init__(self,vocab_size):        # initializing the model with the vocabulary size as an argument. This is necessary because the embedding layer needs to know how many unique words (vocabulary size) it will be working with.init (Constructor)
        super().__init__()
        self.embedding=nn.Embedding(vocab_size,100)     # embedding layer to convert word indices to 100-dimensional vectors.Input: word index and Output: 100-dimensional vector.
        self.lstm=nn.LSTM(100,150,batch_first=True)      # LSTM layer to process the sequence of word vectors and capture temporal dependencies. It takes input of size 100 (the embedding dimension) and outputs a hidden state of size 150 for each time step in the sequence.vocab_size = total number of unique words.super().__init__() initializes PyTorch internals
        self.fc=nn.Linear(150,vocab_size)                # fully connected layer to map
    def forward(self,x):
        embedded=self.embedding(x)   # convert word indices to embeddings    
        intermediate_hidden_states,(final_hidden_state,final_cell_state)=self.lstm(embedded)          # pass the embeddings through the LSTM layer to get the output and hidden state
        output=self.fc(final_hidden_state.squeeze(0))   # pass the final hidden state through the fully connected layer to get the output predictions for the next word in the sequence
        return output
        
        
        
```

```python
model=LSTMModel(len(vocab))   # creating an instance of the LSTMModel class with the vocabulary size as an argument. This initializes the model with the appropriate input and output dimensions based on the vocabulary size.
```

```python
device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')   # setting the device to GPU if available, otherwise CPU
model.to(device)   # moving the model to the specified device (GPU or CPU) for training
```

```python
epochs=50
learning_rate=0.001
criterion=nn.CrossEntropyLoss()   # loss function for multi-class classification
optimizer=torch.optim.Adam(model.parameters(), lr=learning_rate)   # optimizer to update model parameters during training using the Adam optimization algorithm with the specified learning rate
```

```python
# training loop
for epoch in range(epochs):
    total_loss=0
    
    for batch_x, batch_y in dataloader:
        batch_x, batch_y = batch_x.to(device), batch_y.to(device)   # move the input and target batches to the same device as the model (GPU or CPU)
        
        optimizer.zero_grad()   # clear previous gradients before backpropagation
        
        output = model(batch_x)   # forward pass: compute predicted outputs by passing inputs to the model
        
        loss = criterion(output, batch_y)   # calculate the loss between predicted outputs and true labels
        
        loss.backward()   # backward pass: compute gradient of the loss with respect to model parameters
        
        optimizer.step()   # update model parameters using the computed gradients
        
        total_loss += loss.item()   # accumulate the loss for this epoch
        
    print(f'Epoch {epoch+1}, Loss: {total_loss:.4f}')   # print the average loss for this epoch
```
